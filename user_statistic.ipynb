# 3个月内至少购买过一次的用户

```python
action = action.drop('model_id',axis=1)
action = action.drop_duplicates()
action
```

```python
action.time = pd.to_datetime(action.time)
action_sort = action.sort_values(by=['user_id','time'],ascending=[True,True])
action_sort
```

```python
def ifbuy(lst):
    if 4 in pd.unique(lst.type):
        return 1
    else:
        return 0
user_buy = action_sort[['user_id','type']].groupby('user_id').apply(ifbuy)
user_buy.head()
```

```python
user_buy = user_buy.reset_index().rename(columns = {0:'ifbuy'})
user_buy = user_buy[user_buy['ifbuy']==1]['user_id'].reset_index(drop=True)
user_buy
```

```python
from pandas import DataFrame
user_buyed = DataFrame({'user_id':user_buy})
user_buyed
```

```python
user_buyed.to_csv('user_buyed.csv')
```

# 新注册用户


```python
user['user_reg_tm'] = pd.to_datetime(user['user_reg_tm'])
from datetime import date,datetime,timedelta
start = datetime(2016,4,5)
end = datetime(2016,4,15)
def user_newregister(user,start,end):
    return  user[(user['user_reg_tm']>=start)&(user['user_reg_tm']<=end)]
user_newreg = user_newregister(user,start,end)
user_newreg 
```

```python
user_newreg = user_newreg[['user_id']].reset_index(drop=True)
user_newreg
```

融合得到新注册和至少购买一次的用户

```python
user_buyedandnew = pd.merge(user_buyed,user_newreg,on='user_id',how='outer')
user_buyedandnew
```

```python
user_buyedandnew.to_csv('user_buyed_and_newreg_5_15.csv',index=False)
```
